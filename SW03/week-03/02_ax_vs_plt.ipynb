```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 11, 12, 13])
```

The "old" way - pure MATLAB API - NOT recommended!

```python
plt.figure(figsize=(4,1.25))

for i,ax_id in enumerate(['ax1','ax2']):
    print(f'loop is in {ax_id}')
    plt.subplot(1,2,i+1) #subplot is CREATED here
    plt.title(ax_id)
    plt.plot(x+i,y)
    print(f'"get current axis": {plt.gca().get_title()}')
```

"Mixed" way

```python
f, axes = plt.subplots(ncols=2, figsize=(4,1.25)) # all the subplots are created first! 
axes[0].set_title('ax1')
axes[1].set_title('ax2')

for i,ax in enumerate(axes):
    plt.plot(x+i,y)
    print('loop is in {}'.format(ax.get_title()))
    print(f'"get current axis": {plt.gca().get_title()}')

```

The "new" way - recommended!

```python
f, axes = plt.subplots(ncols=2, figsize=(4,1.25))
axes[0].set_title('ax1')
axes[1].set_title('ax2')

for i,ax in enumerate(axes):
    ax.plot(x+i,y)
    print('loop is in {}'.format(ax.get_title()))
    print(f'"get current axis": {plt.gca().get_title()}')
```
